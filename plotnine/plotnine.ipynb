```python
from plotnine import ggplot, aes, geom_point, scale_color_manual, labs
import pandas as pd
```

```python
df = pd.read_csv("../penglings.csv")
```

```python
colors = ["#fba044","#9933cb","#479f9f"]
```

```python
scatter_plot = ggplot(df, aes(x='flipper_length_mm', y='body_mass_g', color='species', size='bill_length_mm')) + \
    geom_point(alpha=0.7) + \
    scale_color_manual(values = colors) + \
    labs(x="Flipper Length (mm)", y="Body Mass (g)")
print(scatter_plot)
```
